```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df[["Survived", "Pclass", "Sex", "Age", "Fare"]]#sadece kullanacağım sütunları seçtim
df.head()
```

```python
x= df.drop("Survived",axis="columns") #survived hariç tüm sütunları değişken olarak seçtim
y=df["Survived"]# hedef -bağımlı- değişkenim
```

```python
from sklearn.preprocessing import LabelEncoder #kategorik verileri (metin tabanlı) sayısal değerlere dönüştürmeye yarar
le_Survived= LabelEncoder()
le_Pclass= LabelEncoder()
le_Sex= LabelEncoder()
le_Age= LabelEncoder()
le_Fare= LabelEncoder()
```

```python
x["Pclass_n"]=le_Pclass.fit_transform(x["Pclass"]) # Pclass sütununn içindeki verileri numeriğe dönüştürüp  x dataframe i üzerinde olusturduğu pclass_n sütununa atıyort
x["Sex_n"]=le_Sex.fit_transform(x["Sex"])
x["Age_n"]=le_Age.fit_transform(x["Age"])
x["Fare_n"]=le_Fare.fit_transform(x["Fare"])
```

```python
x.head()
```

```python
x_n=x.drop(["Parch","SibSp","PassengerId","Pclass", "Sex","Age","Fare","Name","Ticket","Cabin","Embarked"], axis=1 )# eski sütunları dropluyoruz kalanlar bizimdir
```

```python
x_n.head()
```

```python
y
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_n,y)
```

```python
model.score(x_n,y)
```
